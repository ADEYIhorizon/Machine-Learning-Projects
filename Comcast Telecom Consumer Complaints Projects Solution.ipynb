```python
#import the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import the data
complaints = pd.read_csv("Comcast_telecom_complaints_data.csv")
```

```python
# view 5 top rows of the imported data
complaints.head(5)
```

```python
# compute the datetime and create a new column 
complaints["date_time"] = complaints["Date_month_year"] + " " + complaints["Time"]
```

```python
# Set the date_time as index

complaints = complaints.set_index(complaints["date_time"])
```

```python
complaints = complaints.sort_values(by='Date_month_year')
# print(df_sorted)
```
